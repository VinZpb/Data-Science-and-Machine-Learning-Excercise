```python
l2=[41,5,1,3,89,32]                         #Sort Function
print(f"List Sebelum di sort: {l2}")
l2.sort()
print(f"List Setelah di sort: {l2}")

l2=[41,5,1,3,89,32]                         #For Loop
print(f"List Sebelum di sort: {l2}")
l2n=[]
for i2 in range(len(l2)):
    l2n.append(min(l2))
    l2.remove(min(l2))
print(f"List Setelah di sort: {l2n}")
```

```python
d3={0:{"Nama":"Apel","Stok":20,"Harga":10000,"Qty":0},
    1:{"Nama":"Jeruk","Stok":15,"Harga":15000,"Qty":0},
    2:{"Nama":"Anggur","Stok":25,"Harga":20000,"Qty":0}}



while True:
    print("Selamat Datang di Pasar Buah")
    print()
    print("List Menu: ")
    print("1. Menampilkan daftar buah")
    print("2. Menabahkan buah")
    print("3. Mengurangi buah")
    print("4. Membeli buah")
    print("5. Exit Program")
    print()
    in3=int(input(f"Masukkan Angka Menu yang ingin dijalankan: "))
    if in3==1:                                                                                      #1
        print("Daftar Buah")                                                                        #
        print()                                                                                     #
        print(f"{'Index':<8} {'Nama':<10} {'Stok':<8} {'Harga':<8}")                                # Print
        for i31 in range(len(d3)):                                                                  # Daftar
            print(f"{i31:<8} {d3[i31]['Nama']:<10} {d3[i31]['Stok']:<8} {d3[i31]['Harga']:<8}")     #
        print()                                                                                     #

    elif in3==2:                                                                                    #2
        innama=input("Masukkan Nama Buah: ")
        instok=int(input("Masukkan Stok Buah: "))
        inharga=int(input("Masukkan Harga Buah: "))
        d3[len(d3)]={"Nama":innama,"Stok":instok,"Harga":inharga,"Qty":0}
        print("Daftar Buah")                                                                        #
        print()                                                                                     #
        print(f"{'Index':<8} {'Nama':<10} {'Stok':<8} {'Harga':<8}")                                # Print
        for i31 in range(len(d3)):                                                                  # Daftar
            print(f"{i31:<8} {d3[i31]['Nama']:<10} {d3[i31]['Stok']:<8} {d3[i31]['Harga']:<8}")     #
        print()                                                                                     #

    elif in3==3:                                                                                    #3
        print("Daftar Buah")                                                                      #
        print()                                                                                   #
        print(f"{'Index':<8} {'Nama':<10} {'Stok':<8} {'Harga':<8}")                              # Print
        for i31 in range(len(d3)):                                                                # Daftar
            print(f"{i31:<8} {d3[i31]['Nama']:<10} {d3[i31]['Stok']:<8} {d3[i31]['Harga']:<8}")   #
        print()                                                                                   #
        outd3=int(input("Masukkan nomor index yang ingin dihapus: "))
        del d3[outd3]
        import copy
        d3dc=copy.deepcopy(d3)
        c32=0
        for i32 in d3dc:
            d3[c32] = d3.pop(i32)
            c32+=1
        print("Daftar Buah")                                                                      #
        print()                                                                                   #
        print(f"{'Index':<8} {'Nama':<10} {'Stok':<8} {'Harga':<8}")                              # Print
        for i31 in range(len(d3)):                                                                # Daftar
            print(f"{i31:<8} {d3[i31]['Nama']:<10} {d3[i31]['Stok']:<8} {d3[i31]['Harga']:<8}")   #
        print()                                                                                   #

    elif in3==4:                                                                                    #4
        print("Daftar Buah")                                                                        #
        print()                                                                                     #
        print(f"{'Index':<8} {'Nama':<10} {'Stok':<8} {'Harga':<8}")                                # Print
        for i31 in range(len(d3)):                                                                  # Daftar
            print(f"{i31:<8} {d3[i31]['Nama']:<10} {d3[i31]['Stok']:<8} {d3[i31]['Harga']:<8}")     #
        print()                                                                                     #
        while True:
            index34=int(input("Masukan index buah yang ingin dibeli: "))
            Qty34=int(input("Masukan jumlah buah yang ingin dibeli: "))
            if d3[index34]["Stok"]<Qty34:
                print(f"Stok tidak cukup, stock {d3[index34]['Nama']} tinggal {d3[index34]['Stok']}")
                print()
                print("Isi Cart: ")                                                                     #           
                print()                                                                                 #
                print(f"{'Nama':<10} {'Qty':<8} {'Harga':<8}")                                          # Daftar
                for i33 in range(len(d3)):                                                              # Belanja     
                    print(f"{d3[i33]['Nama']:<10} {d3[i33]['Qty']:<8} {d3[i33]['Harga']:<8}")           #
                print()                                                                                 #
            else:
                d3[index34]["Qty"]=Qty34
                print("Isi Cart: ")                                                                     #           
                print()                                                                                 #
                print(f"{'Nama':<10} {'Qty':<8} {'Harga':<8}")                                          # Daftar
                for i33 in range(len(d3)):                                                              # Belanja     
                    print(f"{d3[i33]['Nama']:<10} {d3[i33]['Qty']:<8} {d3[i33]['Harga']:<8}")           #
                print()                                                                                 #
            
            statement4=input("Mau beli yang lain? (ya/tidak)= ")
            if statement4=="ya":
                continue
            elif statement4=="tidak":
                print("Daftar Belanja: ")                                                                                                   #           
                print()                                                                                                                     #
                print(f"{'Nama':<10} {'Qty':<8} {'Harga':<8} {'Total Harga':<15}")                                                          # Daftar
                for i34 in range(len(d3)):                                                                                                  # Belanja     
                    print(f"{d3[i34]['Nama']:<10} {d3[i34]['Qty']:<8} {d3[i34]['Harga']:<8} {d3[i34]['Qty']*d3[i34]['Harga']:<8}")          # Final
                print()                                                                                                                     #
                Total=int()
                for i35 in range(len(d3)):
                    Total+=d3[i35]['Qty']*d3[i35]['Harga']
                print(f"Total yang harus dibayar = {Total}")
                Uang=int(input("Masukan Jumlah Uang:"))
                while Uang<Total:
                    print("Uang tidak mencukupi")
                    Uang=int(input("Masukan Jumlah Uang:"))
                if Uang==Total:
                    print("Terima Kasih")
                else:
                    print("Terima Kasih")
                    print(f"Uang kembalian anda sebesar {Uang-Total}")
                break

    elif in3==5:                                                                                      #5
        break
    else:
        continue

```
