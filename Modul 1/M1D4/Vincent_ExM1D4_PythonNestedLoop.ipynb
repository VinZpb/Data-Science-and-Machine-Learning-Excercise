**Soal 6**

![Soal 6.png](<attachment:Soal 6.png>)

```python
ListBuah=["Apel","Jeruk","Anggur"]
ListBelanja=[0]*3
ListHarga=[10000,15000,20000]
ListStok=[0,7,6]
ListBelanja[0]=int(input(f"Masukkan Jumlah {ListBuah[0]}: "))

for i6 in range(3):
    if ListBelanja[i6]>=ListStok[i6]:
        continue
    else:
        ListBelanja[i6]=int(input(f"Masukkan Jumlah {ListBuah[i6]}: "))
        while ListBelanja[i6]>ListStok[i6]:
            print("Jumlah yang dimasukan terlalu banyak")
            print(f"Stok {ListBuah[i6]} Tinggal: {ListStok[i6]}")
            ListBelanja[i6]=int(input(f"Masukkan Jumlah {ListBuah[i6]}: "))


Total=0
print("")
print("Detail Belanja")
print("")
for i6 in range(3):
    print(f"{ListBuah[i6]} : {ListBelanja[i6]} x {ListHarga[i6]} = {ListBelanja[i6]*ListHarga[i6]}")
    Total+=ListBelanja[i6]*ListHarga[i6]
print("")
print(f"Total : {Total}")
print("")

Uang=int(input("Masukkan Jumlah Uang : "))
while Uang<Total:
    print(f"Uang anda kurang sebesar : {Total-Uang}")
    Uang=int(input("Masukkan Jumlah Uang : "))
if Uang==Total:
    print("Terima Kasih")
else:
    print("Terima Kasih")
    print(f"Uang kembali anda : {Uang-Total}")


```
