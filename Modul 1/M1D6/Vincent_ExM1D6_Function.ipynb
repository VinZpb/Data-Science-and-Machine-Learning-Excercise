**Soal 1**

Buatlah function untuk menghitung luas lingkaran dengan parameter berupa radius

```python
import math
def a1(r):
    print(f"Jari-jari lingkaran {r} mempunyai luas sebesar {round(math.pi*(r)**2,2)}")

r=float(input("Masukkan jari-jari lingkaran: "))
a1(r)
```

```python
def ltarif(lstart,lstop):
    x2=[]
    for i2 in range(lstart,lstop):
        x2.append(8000+5000*i2)
    return x2

print(ltarif(1,6))
```
