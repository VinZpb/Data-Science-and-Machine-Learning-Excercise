```python
txt="Kamu harus menampilkan hanya kata yang berawalan huruf k pada kalimat ini"
txt=txt.split()
for no in txt:
    if no[0]=="k" or no[0]=="K":
        print(no, end=" ")
```

```python
txt3="Tampilkan kata yang memiliki jumlah karakter berjumlah genap pada kalimat ini"
txt3=txt3.split()
for x3 in txt3:
    if len(x3)%2==0:
        print(x3,end=" ")
```

![Soal 4.png](<attachment:Soal 4.png>)

```python
for x4 in range(1,101):
    if x4%3==0 and x4%5==0:
        print(f"{x4} merupakan kelipatan 3 & 5")
    elif x4%3==0:
        print(f"{x4} merupakan kelipatan 3")
    elif x4%5==0:
        print(f"{x4} merupakan kelipatan 5")
    else:
        print(x4)
```
