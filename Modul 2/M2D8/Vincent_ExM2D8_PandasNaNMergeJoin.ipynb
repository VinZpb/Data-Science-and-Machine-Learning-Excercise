```python
#

import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset('mpg')
df
```

1. Tampilkan nilai yang paling sering muncul pada kolom 'origin'

```python
df["origin"].mode()
```

2. Tampilkan rata-rata horsepower

```python
df["horsepower"].mean()
```

3. Tampilkan rata-rata acceleration tiap cylinder

```python
df[["cylinders","acceleration"]].groupby(by="cylinders").mean()
```

4. Tampilkan rata-rata acceleration tiap cylinder dan origin

```python
df[["cylinders","origin","acceleration"]].groupby(by=["cylinders","origin"]).mean()
```

5. Tampilkan rata-rata acceleration tiap cylinder dan origin, kemudian urutkan hasilnya dari acceleration paling tinggi.

```python
df[["cylinders","origin","acceleration"]].groupby(by=["cylinders","origin"]).mean().sort_values(by="acceleration",ascending=False)
```

6. Tampilkan jumlah mobil tiap cylinder

```python
df[["cylinders","name"]].groupby(by="cylinders").count()
```

7. Tampilkan mobil dari japan dengan nilai mpg paling tinggi

```python
df[(df["origin"]=="japan")&(df["mpg"]==df["mpg"].max())][["name"]]
```

8. Displacement adalah jumlah volume total dari seluruh cylinder. Buat kolom baru bernama 'volume_per_cylinder'.

```python
df["volume_per_cylinder"]=df["displacement"]/df["cylinders"]
df
```

9. Tampilkan jumlah mobil keluaran tahun 1975

```python
df["model_year"][df["model_year"]==75].count()
```

10. Buat kolom baru bernama 'origin_brand' berisi nama origin dan brand-nya.

    Contoh:
    - usa chevrolet
    - usa buick
    - europe vw

```python
df["brand"]=df["name"].str.split(" ").str[0]
df["origin_brand"]=df["origin"]+" "+df["brand"]
df.drop(columns="brand",inplace=True)
df
```

11. Apakah ada kolom missing values? Jika ada, tampilkan yang punya missing values!

```python
df.isna().sum()[df.isna().sum()>0]
```

```python
df[df["horsepower"].isna()]
```

```python
df["horsepower"].fillna(value=df["horsepower"].median(),inplace=True)
df
```
