```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('all_cars.csv')
df
```

# **Soal 1**

1. Buat histogram untuk melihat persebaran data pajak mobil (tax)


```python
plt.figure(figsize=(15,4))
sns.histplot(df['tax'],kde=True)
plt.title('Tax Histogram')
plt.ylabel('Frequency')
plt.legend(['Tax Distribution'])
plt.show()
```

# **Soal 2**

2. Buat histogram untuk membandingkan persebaran data harga mobil (price) untuk tiap brand mobil (brand)

```python


fig, axs = plt.subplots(3, 1, figsize=(12, 12))
# plt.title('Tax Histogram')

c2 = 0
for brand2 in df['brand'].unique():
    sns.histplot(data =df[df['brand']==brand2]['price'],kde=True, ax=axs[c2])
    c2+=1



# sns.histplot(data =df[df['brand']=='Audi']['price'],kde=True, ax=axs[1])

# plt.ylabel('Frequency')
# plt.legend(['Tax Distribution'])


# sns.histplot(data =df[df['brand']=='Audi']['price'],kde=True, ax=axs[1])
# plt.title('bbbb')
# plt.ylabel('hhhh')
# plt.legend(['uuuu'])

plt.show()


```

```python
sns.histplot(data=df, x='price',kde=True, hue='brand')
```

# **Soal 3**

3. Buat boxplot untuk melihat persebaran data jarak tempuh (mileage)

```python
sns.boxplot(df,x='mileage')
```

# **Soal 4**

4. Buat boxplot untuk membandingkan persebaran jarak tempuh (mileage) untuk tiap brand mobil (brand)

```python
sns.boxplot(df,x='mileage',y='brand')
```

# **Soal 5**

5. Buat lineplot untuk melihat perkembangan rata-rata harga mobil (price) untuk tiap tahun produksi (year)

```python
plt.figure(figsize=(15,4))
# df['YearTime'] = pd.to_datetime(df['year'])
sns.lineplot(data=df,x='year',y='price', estimator=np.mean)
plt.xticks(df['year'], rotation=(90))
plt.show()
```
